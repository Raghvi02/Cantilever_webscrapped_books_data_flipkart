```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('books.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html)
```

```python
containers=soup.find_all('div' ,{'class':'slAVV4'})
```

```python
# Lists to hold the extracted data
titles = []
ratings = []
prices = []


# Iterate through each book container
for i in containers:
    # Extract title
    a_tag = i.find('a', {'class': 'wjcEIp'})
    if a_tag:
        title = a_tag.text.strip()  # Extract text from the <a> tag and strip whitespace
        titles.append(title)
    else:
        titles.append('Unknown book')
    rating_tag = i.find('div', {'class': '_5OesEi afFzxY'})
    if rating_tag:
        # Assuming 'span id' is incorrect; typically 'id' should be a separate attribute, not within a tag.
        rating_span = rating_tag.find('span')  # Find <span> inside the div
        if rating_span:
            rating_div = rating_span.find('div', {'class': 'XQDdHH'})
            if rating_div:
                rating = rating_div.text.strip()  # Extract text from the <div> tag and strip whitespace
                ratings.append(rating)
            else:
                ratings.append('0')  # No rating div found
        else:
            ratings.append('0')  # No span found
    else:
        ratings.append('3') 
    price_tag= i.find('a',{'class':'DMMoT0'})
    if price_tag:
        price_div=price_tag.find('div',{'class':'hl05eU'})
        if price_div:
            price_div2= price_div.find('div',{'class':'Nx9bqj'})
            if price_div2:
                prices.append(price_div2.text.strip())
            else:
                prices.append(np.nan)
        else:
            prices.append(np.nan)
    else:
            prices.append(np.nan)         
print("ray:", ratings)
```

```python
dict={
    'Title': titles,
    'Price': prices,
    'Rating': ratings
}
```

```python
df=pd.DataFrame(dict)
```
